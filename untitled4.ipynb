## Data preprocessing and feature engineering

### Subtask:
Clean and preprocess the collected data. Extract relevant features for assessing market personality and calculating the mood score. This will likely involve natural language processing for text data and time-series analysis for price and flow data.


**Reasoning**:
Create placeholder dataframes for each data stream, including relevant columns, and perform basic text preprocessing for text-based data.



```python
import pandas as pd
import numpy as np
import string

# 1. Create placeholder data structures
# Financial News
news_data = {
    'timestamp': pd.to_datetime(pd.date_range(end='2025-09-28 07:20:16', periods=100, freq='H')),
    'headline': ['Example news headline about market up', 'Another headline about economy down', 'Market commentary', 'Company earnings report', 'Global market analysis'] * 20,
    'content': ['This is the full content of the news article talking about positive market trends.',
                'Detailed analysis of a market downturn and its causes.',
                'Expert commentary on current market conditions and future outlook.',
                'Summary of a recent company earnings report showing strong performance.',
                'Analysis of global economic factors impacting the market.'] * 20,
    'source': ['Source A', 'Source B', 'Source C', 'Source A', 'Source B'] * 20
}
df_news = pd.DataFrame(news_data)

# Social Media Sentiment
social_media_data = {
    'timestamp': pd.to_datetime(pd.date_range(end='2025-09-28 07:20:16', periods=100, freq='H')),
    'text': ['Market is going up!', 'Feeling bearish today.', 'Neutral sentiment.', 'Very positive outlook.', 'Negative market sentiment.'] * 20,
    'sentiment_score': np.random.uniform(-1, 1, 100)
}
df_social_media = pd.DataFrame(social_media_data)

# Price Behavior Data
price_data = {
    'timestamp': pd.to_datetime(pd.date_range(end='2025-09-28 07:20:16', periods=200, freq='min')),
    'open': np.random.rand(200) * 100 + 1000,
    'high': np.random.rand(200) * 10 + 1005,
    'low': np.random.rand(200) * 10 + 995,
    'close': np.random.rand(200) * 10 + 1000,
    'volume': np.random.randint(1000, 1000000, 200)
}
df_price = pd.DataFrame(price_data)

# Institutional Flow Indicators
institutional_flow_data = {
    'timestamp': pd.to_datetime(pd.date_range(end='2025-09-28 07:20:16', periods=50, freq='D')),
    'buy_volume': np.random.randint(10000, 10000000, 50),
    'sell_volume': np.random.randint(10000, 10000000, 50),
    'net_flow': np.random.uniform(-1000000, 1000000, 50)
}
df_institutional_flow = pd.DataFrame(institutional_flow_data)

# 2. Text preprocessing for Financial News and Social Media Sentiment
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

df_news['content_processed'] = df_news['content'].apply(preprocess_text)
df_social_media['text_processed'] = df_social_media['text'].apply(preprocess_text)

display(df_news.head())
display(df_social_media.head())
display(df_price.head())
display(df_institutional_flow.head())
```

**Reasoning**:
Calculate technical indicators for price data and flow features for institutional flow data.



```python
# 3. Calculate technical indicators for Price Behavior Data
# Calculate 50-minute Moving Average
df_price['50_min_MA'] = df_price['close'].rolling(window=50).mean()

# Calculate Volatility (Standard Deviation of recent price changes)
df_price['price_change'] = df_price['close'].diff()
df_price['volatility'] = df_price['price_change'].rolling(window=50).std()

# 4. Calculate features for Institutional Flow Indicators
# Calculate Flow Strength (e.g., absolute net flow)
df_institutional_flow['flow_strength'] = df_institutional_flow['net_flow'].abs()

# Calculate Flow Imbalance (e.g., difference between buy and sell volume)
df_institutional_flow['flow_imbalance'] = df_institutional_flow['buy_volume'] - df_institutional_flow['sell_volume']

display(df_price.head())
display(df_institutional_flow.head())
```

**Reasoning**:
Combine the processed dataframes into a single dataframe by merging on the timestamp, handling different frequencies by resampling or forward-filling.



```python
# 5. Combine processed features into a single master DataFrame
# Set timestamp as index for merging
df_news = df_news.set_index('timestamp')
df_social_media = df_social_media.set_index('timestamp')
df_price = df_price.set_index('timestamp')
df_institutional_flow = df_institutional_flow.set_index('timestamp')

# Resample institutional flow data to hourly frequency and forward fill
df_institutional_flow_hourly = df_institutional_flow.resample('h').ffill()

# Resample price data to hourly frequency and take the last close price and mean of other features
df_price_hourly = df_price.resample('h').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum',
    '50_min_MA': 'last',
    'price_change': 'sum',
    'volatility': 'last'
})

# Merge the dataframes
master_df = df_price_hourly.merge(df_news[['content_processed']], left_index=True, right_index=True, how='left')
master_df = master_df.merge(df_social_media[['text_processed', 'sentiment_score']], left_index=True, right_index=True, how='left')
master_df = master_df.merge(df_institutional_flow_hourly[['net_flow', 'flow_strength', 'flow_imbalance']], left_index=True, right_index=True, how='left')

# Forward fill missing values after merging
master_df = master_df.ffill()

display(master_df.head())
display(master_df.tail())
```

## Market personality assessment logic

### Subtask:
Implement the logic to analyze the processed data and determine the dominant market personality (Herd Mentality, Contrarian Mode, Efficient Mode, Emotional Mode) based on the defined criteria.


**Reasoning**:
Define the criteria for each market personality and implement the logic to determine the dominant personality type based on the features in the master_df, then store it in a new column.



```python
def determine_market_personality(row):
    """Determines the market personality based on predefined criteria."""

    # Define thresholds and criteria (these are example values and can be adjusted)
    sentiment_threshold_high = 0.5
    sentiment_threshold_low = -0.5
    volatility_threshold_high = master_df['volatility'].quantile(0.75)
    volatility_threshold_low = master_df['volatility'].quantile(0.25)
    flow_strength_threshold_high = master_df['flow_strength'].quantile(0.75)
    flow_imbalance_threshold_high_buy = master_df['flow_imbalance'].quantile(0.75)
    flow_imbalance_threshold_high_sell = master_df['flow_imbalance'].quantile(0.25)

    # Herd Mentality: High social media sentiment score and high flow strength
    if row['sentiment_score'] > sentiment_threshold_high and row['flow_strength'] > flow_strength_threshold_high:
        return 'Herd Mentality'

    # Contrarian Mode: Low social media sentiment score and high flow imbalance (opposite of sentiment)
    if row['sentiment_score'] < sentiment_threshold_low and (row['flow_imbalance'] > flow_imbalance_threshold_high_buy or row['flow_imbalance'] < flow_imbalance_threshold_high_sell):
         return 'Contrarian Mode'

    # Emotional Mode: High volatility and significant news or social media activity
    if row['volatility'] > volatility_threshold_high and (row['sentiment_score'] > sentiment_threshold_high or row['sentiment_score'] < sentiment_threshold_low):
         return 'Emotional Mode'

    # Efficient Mode: Low volatility and neutral sentiment
    if row['volatility'] < volatility_threshold_low and row['sentiment_score'] >= sentiment_threshold_low and row['sentiment_score'] <= sentiment_threshold_high:
        return 'Efficient Mode'

    # Default or Undetermined
    return 'Undetermined'

# Apply the function to determine market personality for each row
master_df['market_personality'] = master_df.apply(determine_market_personality, axis=1)

# Display the updated dataframe with the new column
display(master_df.head())
display(master_df.tail())
```

## Mood calculation logic

### Subtask:
Implement the logic to calculate the market mood score (0-100) by combining the weighted sentiment, price behavior, and institutional flow data as specified.


**Reasoning**:
Define weights for each data stream, normalize the relevant features, calculate the market mood score using the weighted normalized features, scale the score to 0-100, and store it in a new column.



```python
from sklearn.preprocessing import MinMaxScaler

# 1. Define weights for each data stream
weights = {
    'sentiment_score': 0.4,
    'volatility': 0.3,
    'net_flow': 0.3
}

# Ensure weights sum up to 1
total_weight = sum(weights.values())
if total_weight != 1.0:
    weights = {k: v / total_weight for k, v in weights.items()}

# 2. Normalize the relevant features
features_to_normalize = ['sentiment_score', 'volatility', 'net_flow']
scaler = MinMaxScaler()

# Handle potential NaNs before scaling by filling with a placeholder like 0 or mean/median
# For this example, we'll fill NaNs with 0, but a more robust approach might use mean/median
master_df_cleaned = master_df[features_to_normalize].fillna(0)

master_df[features_to_normalize] = scaler.fit_transform(master_df_cleaned)

# 3. Implement the formula to calculate the market mood score
# A higher sentiment score and net flow should generally contribute to a higher mood score.
# Higher volatility might indicate uncertainty, which could lower the mood score, or excitement, which could increase it.
# Let's assume higher volatility slightly decreases the mood score for this example.
master_df['market_mood_score'] = (
    weights['sentiment_score'] * master_df['sentiment_score'] +
    weights['net_flow'] * master_df['net_flow'] +
    weights['volatility'] * (1 - master_df['volatility']) # Inverting volatility for the score calculation
)

# Scale the combined score to a range of 0 to 100
# The current score is in the range of 0 to 1 (since features were scaled to 0-1 and weights sum to 1)
master_df['market_mood_score'] = master_df['market_mood_score'] * 100

# 5. Display the head and tail of the updated master_df
display(master_df.head())
display(master_df.tail())
```

## Emergency scenario handling

### Subtask:
Implement the logic to detect and respond to the defined emergency scenarios (Personality Sudden Shift, Mood Crash, Volatility Spike, News Shock) by adjusting positions or moving to cash.


**Reasoning**:
Define the `handle_emergency_scenarios` function to detect and respond to emergency scenarios, then apply it to the master dataframe and display the results.



```python
import numpy as np

def handle_emergency_scenarios(row, prev_row):
    """
    Detects and responds to emergency scenarios based on market indicators.

    Args:
        row: The current row (representing the current time step).
        prev_row: The previous row (representing the previous time step).

    Returns:
        A tuple containing the detected emergency scenario and the recommended action.
    """
    scenario = "None"
    action = "Maintain strategy"

    # Define thresholds (these are example values and should be calibrated)
    mood_crash_threshold = 20 # Significant drop in mood score
    volatility_spike_threshold_factor = 1.5 # Volatility increases by 50%
    news_sentiment_threshold_negative = -0.7 # Very negative news sentiment
    news_sentiment_threshold_positive = 0.7 # Very positive news sentiment

    # Scenario 1: Personality Sudden Shift
    if prev_row is not None and row['market_personality'] != prev_row['market_personality']:
        scenario = "Personality Sudden Shift"
        action = "Adjust position" # Or "Issue alert", "Re-evaluate strategy"

    # Scenario 2: Mood Crash
    if prev_row is not None and (prev_row['market_mood_score'] - row['market_mood_score']) > mood_crash_threshold:
        scenario = "Mood Crash"
        action = "Move to cash" # Or "Issue alert", "Reduce exposure"

    # Scenario 3: Volatility Spike
    if prev_row is not None and prev_row['volatility'] != 0 and (row['volatility'] / prev_row['volatility']) > volatility_spike_threshold_factor:
        scenario = "Volatility Spike"
        action = "Adjust position" # Or "Reduce position size", "Widen stop-losses"

    # Scenario 4: News Shock (based on sentiment score from social media)
    # Assuming sentiment_score from master_df reflects news/social media combined sentiment
    if row['sentiment_score'] < news_sentiment_threshold_negative or row['sentiment_score'] > news_sentiment_threshold_positive:
         scenario = "News Shock"
         action = "Issue alert" # Or "Re-evaluate position"


    return scenario, action

# Initialize new columns
master_df['emergency_scenario'] = "None"
master_df['emergency_action'] = "Maintain strategy"

# Apply the function to the master_df
# We need to iterate and pass the previous row to the function
for i in range(1, len(master_df)):
    row = master_df.iloc[i]
    prev_row = master_df.iloc[i-1]
    scenario, action = handle_emergency_scenarios(row, prev_row)
    master_df.loc[master_df.index[i], 'emergency_scenario'] = scenario
    master_df.loc[master_df.index[i], 'emergency_action'] = action

# Display the head and tail of the updated master_df
display(master_df.head())
display(master_df.tail())
```
